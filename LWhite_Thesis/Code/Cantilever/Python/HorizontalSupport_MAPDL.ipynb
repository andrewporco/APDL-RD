```python
def HetMatProp(a,b,c,d,c2,CELL,X_start,Y_start):
    ## SC
    mapdl.mp('DENS',2,8862)# kg m^-3
    mapdl.mp('KXX',2,12.397)# W m^-1 C^-1
    mapdl.mp('KYY',2,12.397)# W m^-1 C^-1
    mapdl.mp('KZZ',2,12.397)# W m^-1 C^-1
    mapdl.mp('C',2,421)# J kg^-1 C^-1
    mapdl.mp('PRXY', 2, 0.33)
    mapdl.mp('ALPX', 2, -0.016)      
    mapdl.mp('ALPY', 2, -0.016)     
    mapdl.mp('ALPZ', 2, 0.014)
    mapdl.mp('EX', 2, 7.64E9)
    mapdl.mp('EY', 2, 7.64E9)
    mapdl.mp('EZ', 2, 7.64E9)
    mapdl.mp('GXY', 4, 5.419E+08)
    mapdl.mp('GYZ', 4, 5.419E+08)
    mapdl.mp('GXZ', 4, 5.419E+08)
    
    ## BC
    mapdl.mp('DENS',3,8862)# kg m^-3
    mapdl.mp('KXX',3,24.392)# W m^-1 C^-1
    mapdl.mp('KYY',3,24.392)# W m^-1 C^-1
    mapdl.mp('KZZ',3,24.392)# W m^-1 C^-1
    mapdl.mp('C',3,421)# J kg^-1 C^-1
    mapdl.mp('PRXY', 3, 0.33)
    mapdl.mp('ALPX', 3, -0.016)      
    mapdl.mp('ALPY', 3, -0.016)     
    mapdl.mp('ALPZ', 3, 0.014)      
    mapdl.mp('EX', 3, 3.69E9)
    mapdl.mp('EY', 3, 3.69E9)
    mapdl.mp('EZ', 3, 9.73E9)
    mapdl.mp('GXY', 3, 4.375E09)
    mapdl.mp('GYZ', 3, 4.175E09)
    mapdl.mp('GXZ', 3, 4.175E09)
    ## FC
    mapdl.mp('DENS',4,8862)# kg m^-3
    mapdl.mp('KXX',4,39.43)# W m^-1 C^-1
    mapdl.mp('KYY',4,39.43)# W m^-1 C^-1
    mapdl.mp('KZZ',4,39.43)# W m^-1 C^-1
    mapdl.mp('C',4,421)# J kg^-1 C^-1
    mapdl.mp('PRXY', 4, 0.33)
    mapdl.mp('ALPX', 4, -0.016)      
    mapdl.mp('ALPY', 4, -0.016)     
    mapdl.mp('ALPZ', 4, 0.014)      
    mapdl.mp('EX', 4, 1.34E10)
    mapdl.mp('EY', 4, 1.34E10)
    mapdl.mp('EZ', 4, 1.90E10)
    mapdl.mp('GXY', 4, 1.598E08)
    mapdl.mp('GYZ', 4, 6.873E09)
    mapdl.mp('GXZ', 4, 6.873E09)
    ########## Material Property Assignment ###################
    H = 2*10**-3 #height of cell [mm]
    W = 2*10**-3 #width of cell [mm]
    D = 2*10**-3 #depth of cell [mm]
    num_cellX_SD1 = int(b/W)
    num_cellY_SD1 = int(c/D)
    num_cellZ = int(d/H)
    minX = 0
    minY = 0
    minZ = 0
    i = 0
    for s in range(num_cellZ):
        for k in range(num_cellY_SD1):
            for t in range(num_cellX_SD1):
                x = minX+t*W
                y = minY+k*D            
                z = minZ+s*H
                mapdl.esel('S','CENT','X',x,x+W)
                mapdl.esel('R','CENT','Y',y,y+D)
                mapdl.esel('R','CENT','Z',z,z+H)
#                 print('X',x,x+h,'\n Y',y,y+h,'\n Z',z,z+h)
#                 print(CELL[i])
                if CELL[i] == 'SC':
                    mapdl.mpchg(2,'ALL')
#                     print('changing to SC')
                elif CELL[i] == 'BV':
                    mapdl.mpchg(3,'ALL')
#                     print('changing to BC')
                elif CELL[i] == 'FC':
                    mapdl.mpchg(4,'ALL')
#                     print('changing to FC')
                i+=1
                mapdl.esel('ALL')
    if c2!=0:
        num_cellX_SD2 = int(c2/W)
        num_cellY_SD2 = int(b/D)
        minX = X_start
        minY = Y_start+a
        minZ = 0
#         i = 0
        for s in range(num_cellZ):
            for k in range(num_cellY_SD2):
                for t in range(num_cellX_SD2):
                    x = minX+t*W
                    y = minY+k*D            
                    z = minZ+s*H
                    mapdl.esel('S','CENT','X',x,x+W)
                    mapdl.esel('R','CENT','Y',y,y+D)
                    mapdl.esel('R','CENT','Z',z,z+H)
                    if CELL[i] == 'SC':
                        mapdl.mpchg(2,'ALL')
    #                     print('changing to SC')
                    elif CELL[i] == 'BV':
                        mapdl.mpchg(3,'ALL')
    #                     print('changing to BC')
                    elif CELL[i] == 'FC':
                        mapdl.mpchg(4,'ALL')
    #                     print('changing to FC')
                    i+=1
        
                mapdl.esel('ALL')
```

```python
def Mesh_exact(h):
    mapdl.prep7()
    mapdl.vsel('ALL')
#     mapdl.et(1, 'SOLID279', 0)
    mapdl.et(1, 'SOLID70', 0)
    mapdl.nummrg('ALL')
    mapdl.vglue('ALL')
    mapdl.smrtsize('OFF')
    mapdl.esize(h)
    mapdl.mshkey(0)
    mapdl.mshmid(2)
    mapdl.mshape(1,'3D')
    mapdl.vmesh('ALL')
    mapdl.allsel('ALL')
def Mesh_block(h):
# # Thermal elements
    mapdl.vsel('ALL')
    mapdl.prep7()
    mapdl.et(1, 'SOLID70', 0)
    mapdl.lsel('ALL')
    mapdl.lesize('ALL',h)
    mapdl.allsel('ALL')
    mapdl.vsel('ALL')
    mapdl.vsweep('ALL')
# Structural elements
    mapdl.vsel('ALL')
    mapdl.et(2, 'SOLID185', 0)
    mapdl.lsel('ALL')
    mapdl.lesize('ALL',h)
    mapdl.allsel('ALL')
    mapdl.vsel('ALL')
    mapdl.vsweep('ALL')
```

```python
#### Loadings #####
def Laser(a,b,d,c2,X_start,Y_start,dataL,dataL2):
    
    for II in range(len(dataL)):
        mapdl.nsel('S','LOC','Z',d)
        mapdl.nsel('R','LOC','X',np.around(dataL[II,0]*1e-3,decimals=4))
        mapdl.d('ALL','Temp',dataL[II,1])
    mapdl.nsel('ALL')
    if c2 != 0:
        for II in range(len(dataL2)):
            mapdl.nsel('S','LOC','Z',d)
            mapdl.nsel('R','LOC','Y',np.around(dataL2[II,0]*1e-3,decimals=4))
            mapdl.d('ALL','Temp',dataL2[II,1])
        mapdl.nsel('ALL')
def Tensile(data):
    for II in range(len(data)):
        mapdl.nsel('S','LOC','Z',np.around(data[II,2]*1e-3,decimals=4))
        mapdl.nsel('R','LOC','X',np.around(data[II,0]*1e-3,decimals=4))
        mapdl.nsel('R','LOC','Y',np.around(data[II,1]*1e-3,decimals=4))
        mapdl.f('ALL','FX',data[II,3])
        mapdl.f('ALL','FY',data[II,4])
        mapdl.f('ALL','FZ',data[II,5])
        mapdl.nsel('ALL')
```

```python
def FEM_Thermal(POS,CELL,cell,a,b,c,d,c2,X_start,Y_start,fin,h,begin,MaxDis,MaxPStress):
    a = a/1000
    b = b/1000
    c = c/1000
    d = d/1000
    c2 = c2/1000
    X_start=X_start/1000
    Y_start=Y_start/1000

    Position = POS*10**-3
    cell_path = "\\Para"
    code_path = "\\SourceCode_Backup"
    if begin == 0:
        titleT = f"{os.getcwd()}\XYZForce_Long.txt"
        dataT = np.loadtxt(titleT)
        titleL = f"{os.getcwd()}\Abb_long.txt"
        dataL = np.loadtxt(titleL)
        dataL2 = 0
        mapdl.prep7()
        mapdl.blc4(0,0,(a+b),c,d)
        mapdl.allsel('ALL')
        if c2 != 0:
            mapdl.blc4(X_start,Y_start,c2,(a+b),d)
            titleT = f"{os.getcwd()}\XYZForce_short.txt"
            dataT2 = np.loadtxt(titleT)
            dataT  = np.concatenate((dataT,dataT2),0) 
            titleL = f"{os.getcwd()}\Abb_short.txt"
            dataL2 = np.loadtxt(titleL)
        start_time = time.time()
        Mesh_block(h)
        mapdl.seltol(1e-6)
        mapdl.allsel('ALL')
    ########## Material Properties ###################
    ## Gen
    mapdl.prep7()
    start_time = time.time()
#     if begin == 0:
    mapdl.mp('DENS',1,2670)# kg m^-3
    mapdl.mp('KXX',1,110)# W m^-1 C^-1
    mapdl.mp('KYY',1,110)# W m^-1 C^-1
    mapdl.mp('KZZ',1,110)# W m^-1 C^-1
    mapdl.mp('C',1,915)# J kg^-1 C^-1
    mapdl.mp('PRXY', 1, 0.33)
    mapdl.mp('ALPX', 1, -0.016)      
    mapdl.mp('ALPY', 1, -0.016)     
    mapdl.mp('ALPZ', 1, 0.014)      
    mapdl.mp('EX', 1, 74.0E9)
    mapdl.esel('ALL')
    mapdl.mpchg(1,'ALL')
    mapdl.esel('ALL')
    mapdl.etchg('TTS')
#     HetMatProp(a,b,c,d,c2,CELL,X_start,Y_start,Change,Position)
    HetMatProp(a,b,c,d,c2,CELL,X_start,Y_start)

#     print("Assigning Material --- %s seconds ---" % (time.time() - start_time))
    
#     mapdl.eplot()
    mapdl.run('/SOL')
    mapdl.antype(0)
    if begin == 0:
        start_time = time.time()
        ####### Boundary Conditions ###
        mapdl.autots('ON')            # use automatic time stepping
        mapdl.nsel('S','LOC','Z',0)
        mapdl.d('ALL','ALL')  
        mapdl.allsel('ALL')
        if c2 !=0:
            mapdl.nsel('S','LOC', 'Y', 0)
            mapdl.dsym('SYMM','Y',' ')
        Tensile(dataT)

        mapdl.allsel('ALL')     
        print("Boundary conditions attached--- %s seconds ---" % (time.time() - start_time))

#     solving
    start_time = time.time()
#     print('Solving',begin)
    mapdl.solve()
    mapdl.finish()
    enum,element_stress,enode = mapdl.result.element_stress(-1,principal=True,in_element_coord_sys=False)    
    ElementStress = np.average(np.array(element_stress)[:,:,4],1)
    n_c = 160 #number of clusters
    n_e = int(len(ElementStress)/n_c) #number of elements within each cluster
    p_norm = 8 #penalty
    Sorted = np.sort(ElementStress) #sort high to low
    s_yield = 2.51e8 #yield stress
    vM = np.array([Sorted[x:x+n_e] for x in range(0, len(Sorted),n_e)])#vonMises stress
    nnum,U = mapdl.result.nodal_displacement(-1) #get all nodal displacement
    mag = np.max(LA.norm(U, axis=1)*1e6) #calculate total displacement
    
    S_p_norm = ((np.average((vM/s_yield)**(p_norm),1)))**(1/p_norm) #calculate p-Norm stress
    Stress = np.max(S_p_norm)
#     print("Structural Output--- %s seconds ---" % (time.time() - start_time))

    if Stress<MaxPStress and mag<MaxDis :
        mapdl.prep7()
        mapdl.etchg('STT') #switch to thermal components
        mapdl.run('/SOL')
        mapdl.antype(0)
        if begin == 0: #if iteration = 1
            Laser(a,b,d,c2,X_start,Y_start,dataL,dataL2) #apply temperature distribution
            mapdl.nsel('S','LOC','Z',0)
            mapdl.d('ALL','TEMP',20)
        mapdl.nsel('ALL')
        start_time = time.time()
#         print('Solving',begin)
        mapdl.solve()
        mapdl.finish()
        Node_Z_0 = mapdl.result.mesh.nodes[:,2] #get z component of all nodes
        LIST = mapdl.result.mesh.nnum[np.where(Node_Z_0 == 0)] #find node number where z = 0
        Force,nnode,dof = mapdl.result.nodal_reaction_forces(-1) #get all heat flow
        F = []
        for i in range(len(LIST+1)): #find corresponding node number to heat flow at that node
            rf = Force[np.where(nnode == LIST[i])]
            F.append(rf)
        HEAT = sum(F)
#         print("Thermal Output--- %s seconds ---" % (time.time() - start_time))
    elif begin ==0:
        mapdl.prep7()
        mapdl.etchg('STT')
        mapdl.run('/SOL')
        mapdl.antype(0)
        Laser(a,b,d,c2,X_start,Y_start,dataL,dataL2) #apply temperature distribution
        mapdl.nsel('S','LOC','Z',0) #select nodes at z = 0
        mapdl.d('ALL','TEMP',20)
        mapdl.nsel('ALL')
        start_time = time.time()
#         print('Solving',begin)
        mapdl.solve()
        mapdl.finish()
        Node_Z_0 = mapdl.result.mesh.nodes[:,2]
        LIST = mapdl.result.mesh.nnum[np.where(Node_Z_0 == 0)]
        Force,nnode,dof = mapdl.result.nodal_reaction_forces(-1)
        F = []
        for i in range(len(LIST+1)):
            rf = Force[np.where(nnode == LIST[i])]
            F.append(rf)
        HEAT = sum(F)
#         print("Thermal Output--- %s seconds ---" % (time.time() - start_time))
    else:
        HEAT = False
#         print('Structural Failure')
    if fin == 1:
        mapdl.exit()
    return mag, Stress, HEAT
```
