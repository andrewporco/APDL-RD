```python
    
def FEM_Thermal(CELL,POS,Size,h,begin,MaxDis,MaxPStress):
    cell_path = "\\Para"
    code_path = "\\SourceCode_Backup"
    if begin == 0:
        mapdl.parain('AsymSup','x_t',f"'{os.getcwd()}',SOLIDS,0,0")
        start_time = time.time()

        mapdl.prep7()
        mapdl.seltol(1e-6) #selection tolerance
        mapdl.allsel('ALL')
        Mesh_block(h) #mesh results
    ######### Material Properties ###################
    mapdl.prep7()

    mapdl.mp('DENS',1,8290)# kg m^-3
    mapdl.mp('KXX',1,10.3)# W m^-1 C^-1
    mapdl.mp('KYY',1,10.3)# W m^-1 C^-1
    mapdl.mp('KZZ',1,10.3)# W m^-1 C^-1
    mapdl.mp('C',1,436)# J kg^-1 C^-1
    mapdl.mp('PRXY', 1, 0.319)
    mapdl.mp('ALPX', 1, -0.015)      
    mapdl.mp('ALPY', 1, -0.015)     
    mapdl.mp('ALPZ', 1, 0.02)      
    mapdl.mp('EX', 1, 2.17E11)
      
    start_time = time.time()
    if begin == 0: #it = 0
        InitialMat(CELL,POS,Size)
    else: #it > 0
        HetMatProp(CELL,POS,Size)
    mapdl.etchg('TTS') #change to structural elements
    mapdl.run('/SOL')
    mapdl.antype(0) #static analysis
    if begin == 0:
        start_time = time.time()
        ####### Assign Boundary Conditions ###
        mapdl.autots('ON')            # use automatic time stepping
        mapdl.nsel('S','LOC','Z',0)
        mapdl.d('ALL','ALL')  
        mapdl.allsel('ALL')
        mapdl.input('AA_FLoadAvgSC-bc.txt')
        mapdl.allsel('ALL')     
        print("Structural boundary conditions attached--- %s seconds ---" % (time.time() - start_time))

#     solving
    start_time = time.time()
    mapdl.solve()
    mapdl.finish()

    enum,element_stress,enode = mapdl.result.element_stress(-1,principal=True,in_element_coord_sys=False) #get element stress 
    ElementStress = np.average(np.array(element_stress)[:,:,4],1) #average stress for each element
    n_c = 177 #number of clusters
    n_e = int(len(ElementStress)/n_c) #number of elements within each cluster
    p_norm = 15
    Sorted = np.sort(ElementStress)
    s_yield = 7.12e8#6.12e8 #inconel #7.12e8 = Hanyes 282
    vM = np.array([Sorted[x:x+n_e] for x in range(0, len(Sorted),n_e)]) #vonMises stress
    S_p_norm = ((np.average((vM/s_yield)**(p_norm),1)))**(1/p_norm) #calculate p-Norm stress
    Stress = np.max(S_p_norm)# Get maximum p_norm stress
    nnum,U = mapdl.result.nodal_displacement(-1) #get nodal displacement in x,y,z
    mag = np.max(U[:,2]*1e6) #get maximum nodal displacement in z
    if Stress<MaxPStress and mag<MaxDis : #constraints satisfied?
        mapdl.prep7()
        mapdl.etchg('STT') #change to thermal elements
        mapdl.run('/SOL')
        mapdl.antype(0) #steady-state thermal analysis
        if begin == 0:
            mapdl.input('AA_TLoad.txt')#Input support structure thermal load
            mapdl.nsel('S','LOC','Z',0)
            mapdl.d('ALL','TEMP',200)
        mapdl.nsel('ALL')
        start_time = time.time()
#         print('Solving',begin)
        mapdl.solve()
        mapdl.finish()
        Node_Z_0 = mapdl.result.mesh.nodes[:,2] #get all the z locations of the nodes
        LIST = mapdl.result.mesh.nnum[np.where(Node_Z_0 == 0)] #get node number where z = 0
        Force,nnode,dof = mapdl.result.nodal_reaction_forces(-1) #get the heat flow through entire structure
        F = []
        for i in range(len(LIST+1)): #find the heat flow that corresponds to the area of interest (e.g., where z = 0)
            rf = Force[np.where(nnode == LIST[i])]
            F.append(rf)
        HEAT = sum(F) #add all the nodalheat flow
#         print("Thermal Output--- %s seconds ---" % (time.time() - start_time))
    elif begin ==0: #if i = 0 but does not satify constraints
        mapdl.prep7()
        mapdl.etchg('STT') 
        mapdl.run('/SOL')
        mapdl.antype(0)
        start_time = time.time()
        mapdl.input('AA_TLoad.txt')#Input support structure       
        mapdl.nsel('S','LOC','Z',0)
        mapdl.d('ALL','TEMP',200)
        print("Thermal boundary conditions attached--- %s seconds ---" % (time.time() - start_time))
        mapdl.nsel('ALL')
        start_time = time.time()
        mapdl.solve()
        mapdl.finish()
        #### Plot temperature distribution ##
#         mapdl.result.plot_nodal_temperature(-1, add_text=False,off_screen=True,background='white',text_color='black',cmap='jet')
        
        Node_Z_0 = mapdl.result.mesh.nodes[:,2]
        LIST = mapdl.result.mesh.nnum[np.where(Node_Z_0 == 0)]
        Force,nnode,dof = mapdl.result.nodal_reaction_forces(-1)
        F = []
        for i in range(len(LIST+1)):
            rf = Force[np.where(nnode == LIST[i])]
            F.append(rf)
        HEAT = sum(F)
#         print("Thermal Output--- %s seconds ---" % (time.time() - start_time))
    else: #does not satisfy constraints
        HEAT = False
#         print('Structural Failure')
    return mag, Stress, HEAT
```
